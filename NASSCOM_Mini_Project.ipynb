```python
upper_limit=df['Y house price of unit area'].mean()+3*df['Y house price of unit area'].std()
lower_limit=df['Y house price of unit area'].mean()-3*df['Y house price of unit area'].std()
```

```python
df=df[(df['Y house price of unit area']<upper_limit) & (df['Y house price of unit area']>lower_limit)]
df.shape
```

```python
x=df.iloc[:,1:6].values
y=df.iloc[:,-1].values
```
